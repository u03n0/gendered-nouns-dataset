## CONCATENATE

```python
df_concatenated = pd.concat([fr_df, en_df, es_df, de_df, filtered_pl_df], axis=0, ignore_index=True)
df_concatenated
```

```python
# PL
df_concatenated.loc[df_concatenated['gender'].str.contains('rodzaj męski', case=False, na=False), 'gender'] = 'm'
df_concatenated.loc[df_concatenated['gender'].str.contains('rodzaj żeński', case=False, na=False), 'gender'] = 'f'
# ES
df_concatenated.loc[df_concatenated['gender'].str.contains('Sustantivo_femenino_y_masculino', case=False, na=False), 'gender'] = 'm,f'
df_concatenated.loc[df_concatenated['gender'].str.contains('masculino', case=False, na=False), 'gender'] = 'm'
df_concatenated.loc[df_concatenated['gender'].str.contains('femenino', case=False, na=False), 'gender'] = 'f'
# FR
df_concatenated.loc[df_concatenated['gender'].str.contains('masculin ou féminin', case=False, na=False), 'gender'] = 'm,f'
df_concatenated.loc[df_concatenated['gender'].str.contains('masculin', case=False, na=False), 'gender'] = 'm'
df_concatenated.loc[df_concatenated['gender'].str.contains('féminin', case=False, na=False), 'gender'] = 'f'
# EN
df_concatenated.loc[df_concatenated['gender'].str.contains('f,by sense,m', case=False, na=False), 'gender'] = 'f,m'

df_concatenated
```

```python
filtered_concat = df_concatenated[~(
                   (df_concatenated['noun'].str[0].str.isupper()) |
                   (df_concatenated['noun'].str.isupper()) |
                   (df_concatenated['noun'].str.contains(" ")) |
                   (df_concatenated['noun'].str.contains('\.', na=False)))]
filtered_concat = filtered_concat.reset_index(drop=True)
filtered_concat
```

Duplicates by noun
- Ex: 
- topiaire	m	fr
- topiaire	f	fr

```python
df_duplicates = filtered_concat[filtered_concat.duplicated(subset='noun', keep=False)]
df_duplicates
```

Duplicates by both noun and gender
- topiaire	m	fr
- topiaire	m	fr

```python
df_duplicates_all_columns = filtered_concat[filtered_concat.duplicated(keep=False)]
df_duplicates_all_columns
```

```python
df_no_duplicates = filtered_concat.drop_duplicates(keep='first').reset_index(drop=True)
df_no_duplicates
```

remove nouns with multiple genders

```python

final = df_no_duplicates[df_no_duplicates['gender'].apply(len) == 1]
final
```

multi gendered nouns, saved for later.

```python
multi_final = df_no_duplicates[df_no_duplicates['gender'].apply(len) > 1]
multi_final
```

looks at final df, if noun found in multi_final, we want to remove it 

```python
import pandas as pd

# Assuming df1 and df2 are your DataFrames
# Merge df1 with df2 based on 'noun'
merged_df = pd.merge(final, multi_final[['noun']], on='noun', how='left', indicator=True)

# Filter rows that are only in df1
df3 = merged_df[merged_df['_merge'] == 'left_only'].drop(columns=['_merge'])

# Display the result
df3
```

```python
df3.to_csv('french_nouns_wiktionary.csv')
```
